```python
import numpy as np
import cv2
import matplotlib.pyplot as plt 
```

```python
image_path = "/kaggle/input/my-cat2/fotor-20240418123458.png"
image= cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
plt.imshow(image)
plt.show()
image.shape
```

# Blur 

```python
blur = cv2.GaussianBlur(image, (5,5), 0)
plt.imshow(blur)
plt.show()
image.shape
```

```python
edge= cv2.Canny(blur,20, 150)
plt.imshow(edge)
plt.show()
image.shape
```

```python
contours,hierarchy= cv2.findContours(edge,cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
contour_image= cv2.cvtColor(image, cv2.COLOR_GRAY2BGR)
plt.imshow(contour_image)
plt.show()
image.shape
```

```python
cv2.drawContours(contour_image, contours, -1,(0,255,0),1)
```

```python
plt.figure(figsize=(10,12))

plt.subplot(1,3,1)
plt.imshow(image, cmap= "gray")
plt.title("original")
plt.axis("off")

plt.subplot(1,3,2)
plt.imshow(edge, cmap= "gray")
plt.title("edge Image")
plt.axis("off")

plt.subplot(1,3,3)
plt.imshow(contour_image, cmap= "gray")
plt.title("contour")
plt.axis("off")

plt.show()
```

```python
image = "/kaggle/input/medical/000108 (3).png"
image = cv2.imread(image)
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(gray)
plt.show()
image.shape
```

```python
blur = cv2.GaussianBlur(gray,(5,5), 0)
plt.imshow(blur)
plt.show()
image.shape
```

```python
ret, binary =cv2.threshold(blur, 128,255,cv2.THRESH_BINARY, cv2.THRESH_OTSU)
contours, hierarchy= cv2.findContours(binary,cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
contour_image= image.copy()

cv2.drawContours(contour_image, contours, -1,(0,255,0),1)

gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

fig, axs = plt.subplots(2,2, figsize=(10,10))

axs[0,0].imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
axs[0,0].set_title("original image")
axs[0,0].axis("off")


axs[0,1].imshow(gray, cmap ="gray")
axs[0,1].set_title("gray image ")
axs[0,1].axis("off")




axs[1,0].imshow(binary, cmap ="gray")
axs[1,0].set_title("Binary image ")
axs[1,0].axis("off")

axs[1,1].imshow(cv2.cvtColor(contour_image,cv2.COLOR_BGR2RGB ))
axs[1,1].set_title("contour image ")
axs[1,1].axis("off")

plt.show()
```

```python
image = "/kaggle/input/medical/000108 (3).png"
image = cv2.imread(image)
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

blur = cv2.GaussianBlur(gray,(5,5), 0)
ret, binary =cv2.threshold(blur, 128,255,cv2.THRESH_BINARY + cv2.THRESH_OTSU)
contours, hierarchy= cv2.findContours(binary,cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
contour_image= image.copy()

cv2.drawContours(contour_image, contours, -1,(0,255,0),1)

gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

fig, axs = plt.subplots(2,2, figsize=(10,10))

axs[0,0].imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
axs[0,0].set_title("original image")
axs[0,0].axis("off")


axs[0,1].imshow(gray, cmap ="gray")
axs[0,1].set_title("gray image ")
axs[0,1].axis("off")




axs[1,0].imshow(binary, cmap ="gray")
axs[1,0].set_title("Binary image ")
axs[1,0].axis("off")

axs[1,1].imshow(cv2.cvtColor(contour_image,cv2.COLOR_BGR2RGB ))
axs[1,1].set_title("contour image ")
axs[1,1].axis("off")

plt.show()
```
